```python
#Check if there is any missing values

for col in df.columns:
    missing_percentage = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_percentage*100)))
```

```python
df.info()
```

```python
# Evaluating for Missing Data by using .isnull() or .notnull
# (.isnull()) True: means the value is a missing value while, False: means the value is not a missing value

missing_data = df.isnull()
missing_data
```

```python
# .value_counts() counts the number of 'True' values (missing values)

for column in missing_data.columns.values.tolist():
    print(column)
    
    print (missing_data[column].value_counts())
    print("")   
```

# Data Wrangling

```python
#drop duplicate values

df.drop_duplicates()
```

```python
#replace the '?' symbol with NaN so the dropna() can remove the missing values

df=df.replace('?',np.NaN)
df
```

```python
df=df.dropna()
df
```

```python
for col in df.columns:
    missing_percentage = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_percentage*100)))
```

```python
#Reset Index

df.reset_index(drop=True, inplace=True)
df
```

```python
#List the data type for each column

df.dtypes
```

```python
#Change data type

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df
```

```python
df.head(20)
```

```python
#Create new column 'year_correct' with a correct values for the 'year' column

df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)
df.head(20)
```

```python
df.drop('year', axis = 1, inplace=True)
df.head(20)
```

```python
df.rename(columns={'year_correct':'year'}, inplace=True)
df.head(20)
```

```python
df.dtypes
```

```python
df['year']=df['year'].astype('int64')
```

```python
df.dtypes
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#Sort by 'gross' descendingly

df=df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
#scatter plot of 'bugdet' vs. 'gross' using matplotlib

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show
```

```python
#Looking at correlation for categorical values
#numerized Categorical values

df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
correlation_matrix=df_numerized.corr()
correlation_pairs=correlation_matrix.unstack()
correlation_pairs
```

```python
#take a look at the ones that have a high correlation (> 0.5)

high_correlation=sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
